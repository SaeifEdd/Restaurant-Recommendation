```python
import serpapi
import csv
```

```python
api_key="get an api key"
```

```python
all_places = []

for start in range(0, 500, 20):  # Adjust range as needed
    params = {
        "engine": "google_local",
        "q": "Sousse Restaurants",
        "type": "search",
        "start": start,
        "api_key": api_key
    }

    
    results = serpapi.search(params)
    local_results = results.get("local_results", [])

    if not local_results:
        break  # Stop if no more results

    for place in local_results:
        name = place["title"]
        address = place.get("address")
        place_id = place.get("place_id")
        reviews = place.get("reviews")
        ratings = place.get("rating")
        print(name, "-", address, "-", place_id)
        all_places.append((name, address, place_id, reviews, ratings))
```

```python
#save data got from api
with open("sousse_restaurants_detailed.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Address", "Place_ID", "Reviews", "Rating"])
    writer.writerows(all_places)
```
